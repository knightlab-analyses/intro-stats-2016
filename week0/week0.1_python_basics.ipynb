```python
s1 = {'a', 'b', 'c'}
s2 = {'a', 'd', 'e'}
```

```python
s1 & s2
```

```python
s3 = set(l)
s4 = set(m[0:2])
```

#### if tests

```python
if 'd' in 'abc':
    print('Learn your alphabet.')
elif (2 + 2 == 5):
    print('Sometimes yes.')
else: 
    print('Nothing is true.')    
```

#### while loops

```python
i = 0
while (i < 5):
    print(i)
    i += 1
```

```python
i
```

#### for loops

```python
for x in [0, 1, 2, 3, 4]:
    print(x**2)
```
